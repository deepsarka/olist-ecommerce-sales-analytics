```python
from google.colab import drive
drive.mount('/content/drive')

import pandas as pd

# change this path to match your Drive folder name
path = '/content/drive/MyDrive/olist-data/'

orders = pd.read_csv(path + 'olist_orders_dataset.csv')
order_items = pd.read_csv(path + 'olist_order_items_dataset.csv')
payments = pd.read_csv(path + 'olist_order_payments_dataset.csv')
reviews = pd.read_csv(path + 'olist_order_reviews_dataset.csv')
products = pd.read_csv(path + 'olist_products_dataset.csv')
customers = pd.read_csv(path + 'olist_customers_dataset.csv')
sellers = pd.read_csv(path + 'olist_sellers_dataset.csv')
geolocation = pd.read_csv(path + 'olist_geolocation_dataset.csv')
category_translation = pd.read_csv(path + 'product_category_name_translation.csv')

# quick check that everything loaded correctly
for name, df in [('orders', orders), ('order_items', order_items), ('payments', payments),
                  ('reviews', reviews), ('products', products), ('customers', customers),
                  ('sellers', sellers), ('geolocation', geolocation), ('category_translation', category_translation)]:
    print(name, df.shape)
```

```python
tables = {
    'orders': orders,
    'order_items': order_items,
    'payments': payments,
    'reviews': reviews,
    'products': products,
    'customers': customers,
    'sellers': sellers,
    'geolocation': geolocation,
    'category_translation': category_translation
}

for name, df in tables.items():
    print(f"--- {name} ---")
    display(df.head())
    print()
```

```python
# check missing values in the most important tables first

for name, df in tables.items():
    print(f"--- {name} ---")
    print(df.isnull().sum())
    print()
```

```python
for col in ['product_name_lenght', 'product_description_lenght', 'product_photos_qty']:
    products[col] = products[col].fillna(products[col].median())
```

```python
# Missing value handle

import numpy as np

# Orders — convert dates, coerce missing to NaT (don't drop)
date_cols = ['order_purchase_timestamp', 'order_approved_at',
             'order_delivered_carrier_date', 'order_delivered_customer_date',
             'order_estimated_delivery_date']
for col in date_cols:
    orders[col] = pd.to_datetime(orders[col], errors='coerce')

# Reviews — fill missing text, keep all rows
reviews['review_comment_title'] = reviews['review_comment_title'].fillna('No title')
reviews['review_comment_message'] = reviews['review_comment_message'].fillna('No comment')

# Products — fill missing category, fill numeric with median
products['product_category_name'] = products['product_category_name'].fillna('unknown')
for col in ['product_weight_g', 'product_length_cm', 'product_height_cm', 'product_width_cm']:
    products[col] = products[col].fillna(products[col].median())


print(orders.isnull().sum().sum(), "in orders (expected - undelivered orders)")
print(reviews.isnull().sum().sum(), "in reviews")
print(products.isnull().sum().sum(), "in products")
```

```python
products = products.merge(category_translation, on='product_category_name', how='left')
products['product_category_name_english'] = products['product_category_name_english'].fillna('unknown')
products.head()
```

```python
# Keep only the essential columns from products, drop all the duplicate category columns
products = products[['product_id', 'product_category_name_english']].copy()
products.columns = ['product_id', 'product_category_name_english']
products.head()
```

```python
master = orders.merge(order_items, on='order_id', how='inner')
master = master.merge(products, on='product_id', how='left')
master = master.merge(customers, on='customer_id', how='left')
master = master.merge(sellers, on='seller_id', how='left')

payments_summary = payments.groupby('order_id').agg(
    total_payment_value=('payment_value', 'sum'),
    payment_type=('payment_type', 'first')
).reset_index()
master = master.merge(payments_summary, on='order_id', how='left')

print(master.shape)
master.head()
```

```python
# Delivery delay in days (positive = late, negative = early)
master['delivery_delay_days'] = (master['order_delivered_customer_date'] - master['order_estimated_delivery_date']).dt.days

# Order month/year for trend analysis
master['order_month'] = master['order_purchase_timestamp'].dt.to_period('M').astype(str)
master['order_year'] = master['order_purchase_timestamp'].dt.year

# Total order value = price + freight
master['total_order_value'] = master['price'] + master['freight_value']

master[['order_id', 'delivery_delay_days', 'order_month', 'order_year', 'total_order_value']].head()
```

```python
import os

# create a folder for cleaned data in your Drive
output_path = '/content/drive/MyDrive/olist-data/cleaned/'
os.makedirs(output_path, exist_ok=True)

# export the master table
master.to_csv(output_path + 'master_table.csv', index=False)

# export the individual cleaned tables too (you'll want these for SQL joins practice)
orders.to_csv(output_path + 'orders_clean.csv', index=False)
order_items.to_csv(output_path + 'order_items_clean.csv', index=False)
payments.to_csv(output_path + 'payments_clean.csv', index=False)
reviews.to_csv(output_path + 'reviews_clean.csv', index=False)
products.to_csv(output_path + 'products_clean.csv', index=False)
customers.to_csv(output_path + 'customers_clean.csv', index=False)
sellers.to_csv(output_path + 'sellers_clean.csv', index=False)

print("All files exported to:", output_path)
print(os.listdir(output_path))
```
